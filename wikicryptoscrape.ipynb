```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_cryptocurrencies'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())  # This will print the HTML content in a readable format
```

```python
table = soup.find_all('table')[0]
```

```python
print (table)  # This will print the first table found in the HTML content
```

```python
table_titles_list = table.find_all('th')
```

```python
table_titles_list
```

```python
table_titles = [title.get_text(strip=True) for title in table_titles_list]
```

```python
table_titles
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=table_titles)
```

```python
df
```

```python
column_data = table.find_all('tr')
```

```python
column_data
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print( individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```
